## Why I needed Docker

____

Needed to setup the following stack

1. Website usng nodejs
2. Messaging using Redis
3. Database using MongoDB
4. Orcheatration tool lik Ansible

Each these services must be compatibe with OS version. Some service (or component above) required on dependent library (underlying), another library require another vesion. And when we change one library or dependency, our component or service wont run any more.




With docker we can run each component (1,2,3, 4 above) in a separate containter, with its won libraries and component in same OS, but in seperate environment or "container"




Container are completely isolated environments.The have thier own:

1. Processes or services
2. Networking interfaces
3. Mounts


Difference from VM's, container share same OS kernel. (whereas VM runs its own OS, very bulky).

Some background: Docker uses LXC containers

Since you using OS Kernel, that means you cannot windows docker on a linux machines, Since docker uses the OS's kernel. 

Rememner the main pupose of docker is to CONTAINERIZE APPLICATIONS, not to act as a hypervisors (computer that runs virtual machines) to run VM's, 

Another purpose of Docker is to SHIP and RUN APPLICATIONS.

Difference between Virtual Machines and Docker.
 


Since docker use the same kernel, the containers has less isolation from each other, as they shate the same resources, whereas VM has complete isolation w.r.t to resource sharing. Since the VM does not rely on the underlying kernel or operating system. Also you can have different OS, on same hypervisor, which is not the case of docker


Where do we get these containers:

They can be found in Public Docker Registry called Dockerhub.

## Difference between Images and containers

___

Image is a template to create one or more container.

We can say an image as all the software that runs inside a container.

You can then run a container.


Alor of "products" or self contained applications have been dockerized, meaning that have the best setup of libraries,dependencies, tools, etc for that applications to run optimally. They even provide you with the steps on order to do it.
