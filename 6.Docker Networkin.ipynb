## Docker Fundamentals

__



When you install docker is creates 3 networks automatically:



- bridge
- none
- host

bridge is the default network a container gets attached to.




If you want to manage the networking yourself, you can use the ```network``` parameter, and then you have the following options:





That is we can use the following docker networking commands


```
docker run ubuntu --network=none
docker run ubuntu --network=host
```

Lets look at it more closely...


What is the bridge network is a private internal network created on the host, allowing all the container on the host to communicate with each otherm, and get its own internal ip. In order to acess these containers from the outside world you need to map these ports to docker host



Another way to access the containers is to associated the containers to the host network. Now there is no network isolation between the docker host and container. No port mapping is required here. But now cannot running multiple containers on same host on same port:





With the none network option, the containers are not connected to any network, and cannot be accessed from outside or communicate with other containers.






 




## User defined networks

____


We have investigated the default bridge network.

What if we want to isolate containers within the docker host?

We would like to do the following:



By default docker only creates one internal bridge network.

In order for use to create our own internal network we will need use the following syntax

 
```
docker network create \
    --driver bridge \
    --subnet 182.18.0.0/16
    custom-isolated-network (name of the custom bridge)
```

And then we can list all the networks.


```
docker network ls
```
