## Docker Fundamentals

__

To run a container we used a docker run command. We will not do so in a production environment, where we need to setup a technologystack, that is large number of services that is interacting with each other.

Also instead of running separate docker commands to setup your environment, we rather define our configuration i a Docker Compose file


Docker compose file is a file in yaml format, to define the different services involved in the application

(We will dive deeper into YAML later on)

1. Web
2. Database
3. Messaging
4. Orchestration



And you bring up the entire stack by using the docker-compose up command

```
docker-compose up
```

 




## Docker compose file

____

This is a YAML file. Lets go through the sections:

Start off with a "services" section, with a key and value format. With a dictonary format.


Key = Name of service (web). This can ebe anything: You want to use to name your service.
Value = dictonary. The MINIMUM must have an "image", that must exist on the docker hub or previously built.

These services will be treated as containers with it is ran.

Running the docker-compose up command will bring up the two containers


```
docker-compose up
```

But we have a problem, since an enduser cannot access the application, as we have not mapped the ports.

We can do this at the command line using the docker run command, with the -p parameter. But here we have the configaration in the docker-compose file. So we need to specify it here.


We specify a "ports" property in the docker-compose in the format  ```dockerhost port:container port```. This can be list.


The same can be done in order to do specify our volume mapping:

We achieved this by using the "volumes" property in the format ```dockerhost volumne: container volume```

To stop the container(s) we use docker-compose stop


```
docker-compose stop
```

And to stop and remove the container we use docker-compose down

```
docker-compose down
```
