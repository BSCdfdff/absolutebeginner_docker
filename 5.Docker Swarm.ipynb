## Docker Fundamentals

__

Advanced topic, and out of scope for this beginners series.


Up until now all our containers have been running on a single docker host. This is a point of failure, that is if the host goes down, then all the apps will be unavailable.

We need to have our containers to be replicated across mumtiple docker hosts.




Single host is good for dev test purposes. But not for production.



This is were we use docker swarm. With docker swarm you can manage multiple docker hosts, and will  be treated as a single cluster.




How to set docker swarm up:

We need hosts with docker installed on it. One host will be the master, rest will be workers.

Run ```docker swarm init``` on manager (it will generate a command to be run on the workers), which will allow us now to run the docker swarm commands on the workers. Copy the command and run the "join" command on all the workers. After joining the workers, they become nodes. You are now ready to deploy services on the swarm cluster.





How do we now deploy the servives where we have a docker swarm?

Lets use the same docker-compose file: to deploy multiple instances of a service we add porperty ```deploy```, with the number of replicas.

And to run the compose file we use the docker stack deploy command with the docker-compose file name

```
docker-compose stack deploy -c docker-compose.yml
```

This will deploy 5 stances of services across docker host

There are some additional steps to confugure loadbalancing as well.
